## 資料清理

```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}
df = pd.DataFrame(data)
df
```

```python
df['Company'] = 'TechCorp'
```

```python
df
```

```python
# 新增「年薪」欄位（假設每年有 12 個月薪資）
df['Annual Salary'] = df['Salary'] * 13.5
df
```

```python
# 新增「高薪」欄位（若薪資高於 65000 則標記為 Yes，否則為 No）
def high_salary(salary):
    if salary > 65000:
        return 'Yes'
    else:
        return 'No'

df['High Salary2'] = df['Salary'].apply(high_salary)
df
```

```python
# 刪除「City」與「Department」欄位
df.drop(columns=['High Salary2'], inplace=True)
df
```

```python
df = df.drop(columns=['High Salary'])
df
```

```python
# 使用串列方式新增一筆資料（按照 DataFrame 欄位順序）
df.loc[len(df)] = [111, 'Kevin', 28, 'Austin', 67000, 'IT', 'TechCorp', 360000]
df
```

```python
new_data = [
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
new_df = pd.DataFrame(new_data)
new_df
```

```python
df = pd.concat(  [df, new_df] ,ignore_index=True  )
df
```

```python
# 刪除索引為 2 的資料
df.drop(index=2, inplace=True)
df
```

```python
condition = df['Salary'] < 75000
df = df[condition]
df
```

```python
# 計算每一欄有多少個 NaN
df.isnull().sum()
```

```python
df.isnull()
```

```python
df.loc[4,'Salary'] = pd.NA
df
```

```python
# 用平均數填充 Salary 欄位
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df
```
